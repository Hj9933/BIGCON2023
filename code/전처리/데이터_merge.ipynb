```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv('데이터셋/raw_data_희중.csv', encoding = 'utf-8')
pop = pd.read_csv('행정동별_노인인구수.csv', encoding = 'cp949')
college = pd.read_csv('필지기준_대학교_포함여부.csv', encoding = 'cp949')
```

행정동별 노인인구수 합치기

```python
merge_df = pd.merge(left = raw_data, right = pop, on = ['DATA_CRTR_YM', 'LT_UNQ_NO', 'INDUSTRY_CD'], how = 'left')
merge_df.info()
```

필지별 대학교 포함여부 합치기

```python
merge_df = pd.merge(left = merge_df, right = college, on = ['DATA_CRTR_YM', 'LT_UNQ_NO', 'INDUSTRY_CD'], how = 'left')
merge_df.info()
```
